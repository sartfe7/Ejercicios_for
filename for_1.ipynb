**FUNCIÒN RANGE**::

La función range crea una serie de números consecutivos. La serie que se crea se construye por defecto iniciando en 0 y terminando un número antes del argumento que se le entrega a la función. Cada número de la serie se incrementa en 1. Esta función se declara así.El range genera una secuencia de números que van desde 0 por defecto hasta el número que se pasa como parámetro menos 1. En realidad, se pueden pasar hasta tres parámetros separados por coma, donde el primer es el inicio de la secuencia, el segundo el final y el tercero el salto que se desea entre números.

**5. CÒMO SE REPRESENTA EN UN DIAGRAMA DE FLUJO LA INSTRUCCIÒN FOR**

En un diagrama de flujo, la instrucción repetitiva for se representa típicamente con un rombo o diamante con la palabra "for" escrita dentro. Junto al rombo, se coloca una línea de entrada con la especificación del rango o la colección sobre la cual se iterará. Luego, dentro del rombo, se colocan las acciones o instrucciones que se repetirán en cada iteración del bucle for

```python
# Ejercicio 6: Contar numeros pares e impares

pares = 0
impares = 0

for i in range(20):
    numero = int(input("Introduce un número entero: "))
    if numero % 2 == 0:
        pares += 1
    else:
        impares += 1

# Mostrar resultados
print("Cantidad de números pares:", pares)
print("Cantidad de números impares:", impares)
```

```python
# Ejercicio 7: Contar multiples de 7 y 9 entre 1000 y 5000


multiplos_7 = 0
multiplos_9 = 0
for numero in range(1000, 5001):
    if numero % 7 == 0:
        multiplos_7 += 1
    if numero % 9 == 0:
        multiplos_9 += 1
print(f"Múltiplos de 7: {multiplos_7}")
print(f"Múltiplos de 9: {multiplos_9}")
```

```python
# Ejercico 8: Simular el lanzamiento de N dados y contar ocurrencias

import random

N = int(input("Ingrese el número de dados a lanzar: "))
resultados = {}
for _ in range(N):
    dado = random.randint(1, 6)
    if dado in resultados:
        resultados[dado] += 1
    else:
        resultados[dado] = 1

for cara, veces in resultados.items():
    print
(f"Cara {cara}: {'*' * veces}")
```

**9.1  QUÈ ES UN FACTORIAL?**

El factorial de un entero positivo n, el factorial de n o n factorial se define en principio como el producto de todos los números enteros positivos desde 1 hasta n. Por ejemplo: La operación de factorial aparece en muchas áreas de las matemáticas, particularmente en combinatoria y análisis matemático

```python
# Ejercicio 9: Calcular el factorial de un numero

numero = int(input("Ingrese un número entero : "))
factorial = 1
for i in range(1, numero + 1):
    factorial *= i
print(f"El factorial de {numero} es {factorial}")
```
